# House Price Prediction

### Load Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
house = pd.read_csv('kc_house_data.csv')

house.head()
```

```python
missing_val = house.isnull().sum()
print(missing_val)
```

```python
house.describe()
```

```python
house.info()
house.shape
numeric_df = house.drop(columns=['id', 'date'])
```

```python
corr_matrix = numeric_df.corr()
plt.figure(figsize=(20, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
price_correlation = corr_matrix['price'].abs().sort_values(ascending=False)
print("Correlation with Price:\n", price_correlation)
```

```python
weak_correlation_features = price_correlation[price_correlation < 0.15].index.tolist()

df_cleaned = house.drop(columns=weak_correlation_features)

print("Cleaned Dataset Columns:\n", df_cleaned.columns)
print(df_cleaned.head())
```

```python
df_cleaned = df_cleaned.drop(columns=['id', 'date'])
df_cleaned.head()
```

```python
features = [ 'price',	'bedrooms',	'bathrooms', 'sqft_living', 'floors',	
            'waterfront', 'view', 'grade', 'sqft_above', 'sqft_basement', 'lat', 'sqft_living15']

plt.figure(figsize=(15, 10))
for i, feature in enumerate(features, 1):
    plt.subplot(3, 5, i)
    sns.scatterplot(x=df_cleaned[feature], y=df_cleaned['price'])
    plt.title(f'{feature} vs. Price')
    plt.xlabel(feature)
    plt.ylabel('Price')

plt.tight_layout()
plt.show()
```

```python
corr_matrix = df_cleaned.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix for Cleaned Dataset')
plt.show()
```

```python
numerical_features = [ 'price',	'bedrooms',	'bathrooms', 'sqft_living', 'floors',	
            'waterfront', 'view', 'grade', 'sqft_above', 'sqft_basement', 'lat', 'sqft_living15']
plt.figure(figsize=(15, 10))
for i, feature in enumerate(numerical_features, 1):
    plt.subplot(5, 3, i)
    sns.histplot(df_cleaned[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
